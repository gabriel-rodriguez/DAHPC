## Colormaps

```python
# %load 003/colormaps_reference.py
"""
Reference for colormaps included with Matplotlib.

This reference example shows all colormaps included with Matplotlib. Note that
any colormap listed here can be reversed by appending "_r" (e.g., "pink_r").
These colormaps are divided into the following categories:

Sequential:
    These colormaps are approximately monochromatic colormaps varying smoothly
    between two color tones---usually from low saturation (e.g. white) to high
    saturation (e.g. a bright blue). Sequential colormaps are ideal for
    representing most scientific data since they show a clear progression from
    low-to-high values.

Diverging:
    These colormaps have a median value (usually light in color) and vary
    smoothly to two different color tones at high and low values. Diverging
    colormaps are ideal when your data has a median value that is significant
    (e.g.  0, such that positive and negative values are represented by
    different colors of the colormap).

Qualitative:
    These colormaps vary rapidly in color. Qualitative colormaps are useful for
    choosing a set of discrete colors. For example::

        color_list = plt.cm.Set3(np.linspace(0, 1, 12))

    gives a list of RGB colors that are good for plotting a series of lines on
    a dark background.

Miscellaneous:
    Colormaps that don't fit into the categories above.

"""
import numpy as np
import matplotlib.pyplot as plt


cmaps = [('Sequential',     ['Blues', 'BuGn', 'BuPu',
                             'GnBu', 'Greens', 'Greys', 'Oranges', 'OrRd',
                             'PuBu', 'PuBuGn', 'PuRd', 'Purples', 'RdPu',
                             'Reds', 'YlGn', 'YlGnBu', 'YlOrBr', 'YlOrRd']),
         ('Sequential (2)', ['afmhot', 'autumn', 'bone', 'cool', 'copper',
                             'gist_heat', 'gray', 'hot', 'pink',
                             'spring', 'summer', 'winter']),
         ('Diverging',      ['BrBG', 'bwr', 'coolwarm', 'PiYG', 'PRGn', 'PuOr',
                             'RdBu', 'RdGy', 'RdYlBu', 'RdYlGn', 'Spectral',
                             'seismic']),
         ('Qualitative',    ['Accent', 'Dark2', 'Paired', 'Pastel1',
                             'Pastel2', 'Set1', 'Set2', 'Set3']),
         ('Miscellaneous',  ['gist_earth', 'terrain', 'ocean', 'gist_stern',
                             'brg', 'CMRmap', 'cubehelix',
                             'gnuplot', 'gnuplot2', 'gist_ncar',
                             'nipy_spectral', 'jet', 'rainbow',
                             'gist_rainbow', 'hsv', 'flag', 'prism'])]


nrows = max(len(cmap_list) for cmap_category, cmap_list in cmaps)
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

def plot_color_gradients(cmap_category, cmap_list):
    fig, axes = plt.subplots(nrows=nrows)
    fig.subplots_adjust(top=0.95, bottom=0.01, left=0.2, right=0.99)
    axes[0].set_title(cmap_category + ' colormaps', fontsize=14)

    for ax, name in zip(axes, cmap_list):
        ax.imshow(gradient, aspect='auto', cmap=plt.get_cmap(name))
        pos = list(ax.get_position().bounds)
        x_text = pos[0] - 0.01
        y_text = pos[1] + pos[3]/2.
        fig.text(x_text, y_text, name, va='center', ha='right', fontsize=10)

    # Turn off *all* ticks & spines, not just the ones with colormaps.
    for ax in axes:
        ax.set_axis_off()

for cmap_category, cmap_list in cmaps:
    plot_color_gradients(cmap_category, cmap_list)

plt.show()
```

```python
# %load 003/two_scales.py # Ejemplo: grfica con dos ejes
#!/usr/bin/env python
"""

Demonstrate how to do two plots on the same axes with different left
right scales.


The trick is to use *2 different axes*.  Turn the axes rectangular
frame off on the 2nd axes to keep it from obscuring the first.
Manually set the tick locs and labels as desired.  You can use
separate matplotlib.ticker formatters and locators as desired since
the two axes are independent.

This is achieved in the following example by calling the Axes.twinx()
method, which performs this work. See the source of twinx() in
axes.py for an example of how to do it for different x scales. (Hint:
use the xaxis instance and call tick_bottom and tick_top in place of
tick_left and tick_right.)

The twinx and twiny methods are also exposed as pyplot functions.

"""

import numpy as np
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots( figsize=(15,5) )
t = np.arange(0.01, 10.0, 0.01)
s1 = np.exp(t)
ax1.plot(t, s1, 'b-')
ax1.set_xlabel('time (s)')
# Make the y-axis label and tick labels match the line color.
ax1.set_ylabel('exp', color='b')
for tl in ax1.get_yticklabels():
    tl.set_color('b')


ax2 = ax1.twinx()
s2 = np.sin(2*np.pi*t)
ax2.plot(t, s2, 'r.')
ax2.set_ylabel('sin', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')
plt.show()

```

```python
# Example: move the axes to the middle of the plot
x = np.linspace( -np.pi, np.pi, 500, endpoint=True )
y = np.sin(x)

plt.plot( x, y )
ax = plt.gca()

# The plot has, by default, four borders or "spines": top, bottom, left, and right.
# We hide two of them and move the other two.
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

# We need to "move" the ticks in the hidden axes so that they are not left floating on the border of the figure.
ax.xaxis.set_ticks_position( 'bottom' )
ax.yaxis.set_ticks_position( 'left' )
```

# 2D histograms

```python
import scipy.stats

# We create a simple histogram
mu, sigma = 200, 25
x = mu + sigma * np.random.randn(10000)

# We use histtype="stepfilled"
n, bins, patches = plt.hist(x, 50, histtype='stepfilled')
plt.setp(patches, 'facecolor', 'g', 'alpha', 0.75)

# Adds a line showing a normal distribution
y = scipy.stats.norm.pdf( bins, mu, sigma)
l = plt.plot(bins, y, 'k--', linewidth=1.5)
```

```python
# We create a histogiram providing the bins limits (not equispaced)
bins = [100,125,150,160,170,180,190,200,210,220,230,240,250,275,300]

# We use histtype="bar"
n, bins, patches = plt.hist(x, bins, histtype='bar', rwidth=0.8)
```

```python
# Cumulative histogram
n, bins, patches = plt.hist(x, 50, histtype='step', cumulative=True)

# Adds a line showing the expected normal cumulative distribution
y = scipy.stats.norm.pdf( bins, mu, sigma).cumsum()
y /= y[-1]
l = plt.plot(bins, y, 'k--', linewidth=1.5)
```

```python
# We create a second set of data with a less typical distribution
sigma2 = 15.
x = mu + sigma2*np.random.randn(10000)

# We use histtype="step"
n, bins, patches = plt.hist(x, bins=bins, histtype='step', cumulative=True)

# We add a line showing the expected normal distribution
y = scipy.stats.norm.pdf( bins, mu, sigma2).cumsum()
y /= y[-1]
l = plt.plot(bins, y, 'r--', linewidth=1.5)

# We draw an inverted cumulative histogram on the same plot
n, bins, patches = plt.hist(x, bins=bins, histtype='step', cumulative=-1)


plt.grid(True) # Activate the grid (shows ticks on the canvas)
plt.ylim(0, 1.05)
```

```python
# It is possible to represent the histograms of multiple data series in parallel on the same plot
# We create a set of data representing three different series of one thousand elements each
x = mu + sigma*np.random.randn(1000,3)

# Overwrite the default colors using "color="
n, bins, patches = plt.hist(x, 10, histtype='bar',
                            color=['crimson', 'burlywood', 'chartreuse'],
                            label=['Crimson', 'Burlywood', 'Chartreuse'])

plt.legend() # Enable legend
```

```python
# When representing multiple series at the same time, we can stack them
n, bins, patches = plt.hist(x, 10, histtype='bar', stacked=True,
                            color = ['crimson', 'burlywood', 'chartreuse'],
                            label=['Crimson', 'Burlywood', 'Chartreuse'])
plt.legend()
```

```python
# We can also create stacked histograms using histtype="stepfilled" o "step"
n, bins, patches = plt.hist(x, 10, histtype='stepfilled', stacked=True, color=['crimson', 'burlywood', 'chartreuse'],
                            label=['Crimson', 'Burlywood', 'Chartreuse'] )
plt.legend()
```

```python
# Finally, we build a multiple histogram with datasets of different sizes
x0 = mu + sigma*np.random.randn(10000)
x1 = mu + sigma*np.random.randn(7000)
x2 = mu + sigma*np.random.randn(3000)

# And we use the weight parameter to give the first half of the elements half of the weight of the elements in the second half
w0 = np.ones_like(x0)
w0[:len(x0)//2] = 0.5
w1 = np.ones_like(x1)
w1[:len(x1)//2] = 0.5
w2 = np.ones_like(x2)
w2[:len(x2)//2] = 0.5

n, bins, patches = plt.hist( [x0,x1,x2], 10, weights=[w0, w1, w2], histtype='bar')
```

# Example: filling areas in between lines

```python
x = np.arange( 0.0, 2, 0.01 )

# We build two different sinusoidal waves
y1 = np.sin( 2*np.pi*x )
y2 = 1.2*np.sin( 4*np.pi*x )

plt.plot( x, y1, x, y2, color='k' )

# Filling
plt.fill_between( x, y1, y2, where=y2 >= y1, facecolor='darkblue', interpolate=True )
plt.fill_between( x, y1, y2, where=y2 <= y1, facecolor='deeppink', hatch="/", interpolate=True )

plt.title( 'Filling in between lines' )
```

# Example: contour plot of $f(x,y) = (1 - (x^2 + y^2)) \cdot e^\frac{-y^3}{3}$

```python
import matplotlib.cm as cm

def f( x, y ):
    return (1 - (x**2 + y**2)) * np.exp( -y ** 3 / 3 )

x = np.arange( -1.5, 1.5, 0.1 )
y = np.arange( -1.5, 1.5, 0.1 )
X,Y = np.meshgrid( x, y )
Z = f( X, Y )

# number of isolines
N = np.arange( -1, 1.5, 0.3 )

# We draw the contour plot, labels, and colormap legend
plt.figure( figsize=(8,7) )
CS = plt.contour( Z, N, linewidths=2, cmap=cm.jet )
plt.clabel( CS, inline=True, fmt="%1.1f", fontsize=10 )
plt.colorbar( CS )
```

# Example: plots in polar coordinates

```python
max_value = 20 # Maximum value in the plot
N = 20 # Number of points
colormap = lambda r: cm.Set2( float(r)/max_value ) # colormap accepts values in [0,1]

fig = plt.figure( figsize=(7,7) )
ax = fig.add_axes( [0.2, 0.2, 0.7, 0.7], polar=True )

theta = np.arange( 0.0, 2*np.pi, 2*np.pi/N )
radii = max_value*np.random.rand( N )
width = np.pi / 4 * np.random.rand( N )
bars = ax.bar( theta, radii, width=width, bottom=0.0 )
for r, bar in zip( radii, bars ):
    bar.set_facecolor( colormap(r) )
    bar.set_alpha( 0.6 )
```

# Example: 3D plots

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import random

mpl.rcParams['font.size'] = 10
fig = plt.figure( figsize=(7,7) )
ax = fig.add_subplot( 111, projection='3d' ) # We get an Axis3D object

for z in [2011, 2012, 2013, 2014]:
    xs = range(1,13)
    ys = 1000 * np.random.rand(12)
    
    color = plt.cm.Set2( random.choice( range( plt.cm.Set2.N ) ) )
    ax.bar( xs, ys, zs=z, zdir="y", color=color, alpha=0.8 )
    
ax.xaxis.set_major_locator( mpl.ticker.FixedLocator(xs) )
ax.yaxis.set_major_locator( mpl.ticker.FixedLocator(ys) )
ax.set_xlabel( "Month" )
ax.set_ylabel( u"Year" )
ax.set_zlabel( u"Net sales ()" )
```
